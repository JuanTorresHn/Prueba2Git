````python
# Exercise - create a function that, given a dataframe
# and a column name, prints out the number of null values
# in that column
def print_num_nulls(df, col_name):
    print(f'Column {col_name} has {df[col_name].isnull().sum()} null values')

print_num_nulls(df_orders, 'MAIN')
````

````python
# We can do this for every column in the dataframe by looping
# through df_orders' columns.
for i in df_orders:
    print_num_nulls(df_orders, i)
````
